```python
# Import Pandas Library
import pandas as pd

# Load Excel Dataset
df = pd.read_excel("Dataset for Data Analytics.xlsx")

# Display Records
df.head()
```

```python
#Dataset Overview
#Check number of rows, columns and data types

print("Dataset Shape:", df.shape)
print("\nColumn Names:")
print(df.columns)

print("\nDataset Information:")
print(df.info())
```

```python
#Identify Missing Values
print(df.isnull().sum())
```

```python
#Handle Missing Values
df['CouponCode'] = df['CouponCode'].fillna('No Coupon')

print(df.isnull().sum())
```

```python
#Check Duplicate Records
duplicates = df.duplicated().sum()

print("Duplicate Records:", duplicates)
```

```python
#Remove Duplicate Records
df = df.drop_duplicates()

print("Duplicate Records Removed")
```

```python
#Check Duplicate Order IDs
duplicate_orderid = df.duplicated(
    subset=['OrderID']
).sum()

print("Duplicate Order IDs:",
      duplicate_orderid)
```

```python
# Convert date to DD-MM-YYYY format

df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%d-%m-%Y')

# Check output
print(df['Date'].head())
```

```python
#Validate TotalPrice
df['CalculatedPrice'] = (
    df['Quantity']
    * df['UnitPrice']
)

print(df[['Quantity',
          'UnitPrice',
          'TotalPrice',
          'CalculatedPrice']].head())
```

```python
#Find Incorrect Prices
wrong_price = df[
    round(df['CalculatedPrice'], 2)
    !=
    round(df['TotalPrice'], 2)
]

print(wrong_price)
```

```python
#Final Dataset Validation
print(df.isnull().sum())
```

```python
df["PaymentMethod"].unique()
```

```python
if 'CalculatedPrice' in df.columns:
    df.drop('CalculatedPrice', axis=1, inplace=True)

df.to_excel("Cleaned_Dataset.xlsx", index=False)
```

```python
#Save Cleaned Dataset
df.to_excel(
    "Cleaned_Dataset.xlsx",
    index=False
)

print("Cleaned Dataset Saved Successfully")
```
